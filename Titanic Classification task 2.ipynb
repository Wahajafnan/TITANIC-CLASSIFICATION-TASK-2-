```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pdf=pd.read_csv("C:\\Users\\wahaj\\Downloads\\Titanic\\Titanic-Dataset.csv")
df=pdf
```

```python
df
```

```python
#checking na vales
df.isnull().sum()
```

```python
#replacing na with their mean values
df['Age'].fillna((df['Age'].mean().astype(int)),inplace=True)
```

```python
#again checking na vales
df.isnull().sum()
```

```python
df['Cabin'].isnull().sum()/len(df)
```

```python
df.drop(['Cabin'],axis=1,inplace=True)
df.dropna(inplace=True)
```
